```python
import soundfile
from espnet2.bin.tts_inference import Text2Speech

text2speech = Text2Speech.from_pretrained('mio/amadeus')

print('TTS model loaded!')
```

```python
from IPython.display import HTML, Audio
from IPython.core.display import display

text = 'どうでもいいような 夜だけどときめき 色めきと君も "踊ろう"どうでもいいような 夜だけど二人刻もう。' #@param {type:"string"}
input_audio_path = '/content/out.wav' #@param {type:"string"}

speech = text2speech(text)['wav']
soundfile.write('/content/out.wav', speech.numpy(), text2speech.fs, 'PCM_16')

display(Audio(input_audio_path))
```
